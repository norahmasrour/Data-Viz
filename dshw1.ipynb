```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Create a directory for output CSV files
output_csv_dir = "bernoulli_simulation_results"
os.makedirs(output_csv_dir, exist_ok=True)

# Parameters for the simulation
observations_list = [10, 100, 1000, 10000]
runs_list = [10, 100, 1000, 10000]
probability = 0.7

# Function to perform Bernoulli simulations
def perform_bernoulli_simulation(observations, runs, p):
    sample_means = []
    for _ in range(runs):
        # Simulate Bernoulli trials and compute the mean
        sample = np.random.binomial(1, p, observations)
        sample_means.append(np.mean(sample))
    return sample_means

# Perform simulations and save results as CSV files
for observations in observations_list:
    for runs in runs_list:
        # Perform the simulation
        sample_means = perform_bernoulli_simulation(observations, runs, probability)

        # Create a DataFrame and save to CSV
        df = pd.DataFrame(sample_means, columns=["Sample Mean"])
        filename = f"bernoulli_means_obs_{observations}_runs_{runs}.csv"
        filepath = os.path.join(output_csv_dir, filename)
        df.to_csv(filepath, index=False)

        print(f"Saved: {filename}")

# Visualization of the sample mean distributions
plt.figure(figsize=(20, 15))
plot_index = 1
for observations in observations_list:
    for runs in runs_list:
        # Load the dataset
        filename = f"bernoulli_means_obs_{observations}_runs_{runs}.csv"
        filepath = os.path.join(output_csv_dir, filename)
        df = pd.read_csv(filepath)

        # Plot the histogram
        plt.subplot(len(observations_list), len(runs_list), plot_index)
        plt.hist(df["Sample Mean"], bins=30, alpha=0.7, edgecolor='black')
        plt.title(f"Obs: {observations}, Runs: {runs}")
        plt.xlabel("Sample Mean")
        plt.ylabel("Frequency")
        plot_index += 1

plt.tight_layout()
plt.show()



# Commentary:
# As the number of observations increases, the distribution of the sample means becomes narrower and more centered around the true probability (p=0.7).
# Increasing the number of runs for a fixed number of observations does not change the distribution's spread but provides a smoother histogram representation.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load the dataset
file_path = 'HW1_dataset1.csv'
data = pd.read_csv(file_path)

# Extract the feature
x = data['X']

# 1. Visualization: Plot a histogram
plt.figure(figsize=(8, 6))
plt.hist(x, bins=15, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Histogram of X')
plt.xlabel('X')
plt.ylabel('Density')
plt.grid(alpha=0.3)
plt.show()

# 2. Fit a Normal distribution to the data
mu, std = norm.fit(x)

# 3. Overlay the fitted distribution on the histogram
xmin, xmax = x.min(), x.max()
x_range = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x_range, mu, std)

plt.figure(figsize=(8, 6))
plt.hist(x, bins=15, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')
plt.plot(x_range, pdf, 'k', linewidth=2, label=f'Normal Fit ($\\mu={mu:.2f}, \\sigma={std:.2f}$)')
plt.title('Histogram of X with Fitted Normal Distribution')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Commentary:
# The data for X appears to fit a Normal distribution based on the shape of the histogram.
# This assumption is supported by the bell-shaped curve of the histogram and its peak.
# The estimated parameters are:
# - Mean (mu): The central value of the distribution.
# - Standard deviation (sigma): The spread of the distribution.
# The overlay of the Normal distribution aligns well with the histogram, justifying the model choice.
```
